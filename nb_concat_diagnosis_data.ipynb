```python
import pandas as pd
```

```python
diagnosis_p1 = pd.read_csv("./attempt_3_data/diagnosis_p1_std_model_full.csv")
diagnosis_p2 = pd.read_csv("./attempt_3_data/diagnosis_p2_std_model_full.csv")
```

```python
full_data = pd.concat([diagnosis_p1, diagnosis_p2["euph_response_diagnosis_p2"]], axis=1)
```

```python
full_data.head()
```

```python
import ipywidgets as widgets
from IPython.display import display

row_num = 15
text1 = full_data.iloc[row_num]["euph_response"]
text2 = full_data.iloc[row_num]["euph_response_diagnosis_p1"]
text3 = full_data.iloc[row_num]["euph_response_diagnosis_p2"]

box_layout = widgets.Layout(
    width='33%', 
    height='600px', 
    border='1px solid #ccc', 
    overflow_y='scroll',
    padding='10px'
)

style = 'style="white-space: pre-wrap; word-break: break-all;"'

widget1 = widgets.HTML(value=f"<pre {style}>{text1}</pre>", layout=box_layout)
widget2 = widgets.HTML(value=f"<pre {style}>{text2}</pre>", layout=box_layout)
widget3 = widgets.HTML(value=f"<pre {style}>{text3}</pre>", layout=box_layout)

display(widgets.HBox([widget1, widget2, widget3]))
```
